```python
import numpy as np
import pandas as pd
import os
```

```python
from  sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,log_loss
```

```python
star_ds = pd.read_csv('Dataset/6 class csv.csv')
star_ds.head()
```

```python
star_dum = pd.get_dummies(star_ds,drop_first = True)
star_dum.head()
```

```python
x = star_dum.drop('Star type',axis = 1)
y = star_dum['Star type']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 2022,stratify = y)
```

```python
scaler = StandardScaler()
acc = []
loss = []
for n in range(1,30):
    knn = KNeighborsClassifier(n_neighbors = n)
    pipe = Pipeline([('scl',scaler),('knn',knn)])
    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)
    y_pred_prob = pipe.predict_proba(x_test)
    print(f'k = {n}')
    print(accuracy_score(y_test,y_pred))
    acc.append(accuracy_score(y_test,y_pred))
    print(log_loss(y_test,y_pred_prob))
    loss.append(log_loss(y_test,y_pred_prob))
    print("_"*40)
print(f'Best accuracy score is {np.max(acc)} at k = {(np.argmax(acc)+1)} and \nbest log loss is {np.min(loss)} at k = {(np.argmin(loss)+1)}')
```
